```python
import pandas as pd 
import requests 
```

```python
# read a csv file 
df = pd.read_csv("1-4000_only_ad.csv", index_col=0)
```

```python
df
```

```python
# add http or https to the urls in df.urls
def add_http(url):
    if url.startswith("http"):
        return url
    else:
        return "https://" + url
    
df["url"] = df["url"].apply(add_http)
```

```python
# we have to send a request to the url and check if the status code is 200 
# also use multi-threading to speed up the process and use 100 workers 
# save the results in a new column called "status_code"
# use tqdm to show the progress bar
from tqdm import tqdm
from concurrent.futures import ThreadPoolExecutor
def get_status_code(url):
    try:
        response = requests.get(url, allow_redirects=True, timeout=5)
        return response.status_code
    except:
        # if there is error than return error
        return "error"
    
with ThreadPoolExecutor(max_workers=200) as executor:
    df["status_code"] = list(tqdm(executor.map(get_status_code, df["url"]), total=len(df)))
    

```
